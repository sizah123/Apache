```python


# Load the dataset
#file_path = '/mnt/data/fetalHealth.csv'
#data = pd.read_csv(file_path)

# Data Preprocessing (normalization)


# Function to calculate quantization and topological errors
def evaluate_som(som, data_scaled):
    quantization_error = som.quantization_error(data_scaled)
    topological_error = som.topographic_error(data_scaled) #The function name is topographic_error
    return quantization_error, topological_error

# Experiment with different SOM parameters
som_params = {
    'grid_sizes': [(10, 10), (20, 20), (30, 30)],  # Different grid sizes
    'learning_rates': [0.1, 0.5, 0.01],  # Different learning rates
    'sigmas': [1.0, 2.0, 0.5]  # Different neighborhood sizes
}

best_som = None
best_params = {}
lowest_qe = float('inf')  # Initialize lowest quantization error

# Train SOMs with different architectures
for grid_size in som_params['grid_sizes']:
    for learning_rate in som_params['learning_rates']:
        for sigma in som_params['sigmas']:
            print(f'Training SOM with grid_size={grid_size}, learning_rate={learning_rate}, sigma={sigma}')

            # Initialize and train the SOM
            som = MiniSom(x=grid_size[0], y=grid_size[1], input_len=data_scaled.shape[1], sigma=sigma, learning_rate=learning_rate)
            som.random_weights_init(data_scaled)
            som.train_random(data_scaled, 1000)  # Train for 1000 iterations

            # Evaluate the SOM
            qe, te = evaluate_som(som, data_scaled)
            print(f'Quantization Error: {qe}, Topological Error: {te}')

            # Keep track of the best performing SOM based on quantization error
            if qe < lowest_qe:
                lowest_qe = qe
                best_som = som
                best_params = {
                    'grid_size': grid_size,
                    'learning_rate': learning_rate,
                    'sigma': sigma,
                    'quantization_error': qe,
                    'topological_error': te
                }
```

```python
# Print the best SOM configuration
print("Best SOM Configuration:")
print(best_params)

# Visualize the U-Matrix for the best SOM
plt.figure(figsize=(10, 8))
plt.pcolor(best_som.distance_map().T, cmap='coolwarm')  # U-Matrix
plt.colorbar()
plt.title('Best SOM U-Matrix')
plt.show()

# Visualize Component Maps for each feature
plt.figure(figsize=(20, 15))
for i, feature in enumerate(data.columns[:-1]):
    plt.subplot(5, 5, i+1)
    plt.title(feature)
    plt.imshow(best_som.get_weights()[:, :, i].T, cmap='coolwarm', aspect='auto')
plt.tight_layout()
plt.show()

# Quantization Error and Topological Error of the best SOM
print(f"Quantization Error: {best_params['quantization_error']}")
print(f"Topological Error: {best_params['topological_error']}")
```

```python
#question 5
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, _tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder

# Function to extract decision rules from a decision tree
def tree_rules(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    paths = []
    def recurse(node, path):
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            left_path = f"{name} <= {threshold:.3f}"
            right_path = f"{name} > {threshold:.3f}"
            recurse(tree_.children_left[node], path + [left_path])
            recurse(tree_.children_right[node], path + [right_path])
        else:
            # Leaf node: contains a prediction (i.e., the final class/label)
            leaf_class = np.argmax(tree_.value[node])
            paths.append((path, leaf_class))

    recurse(0, [])
    return paths
# Evaluate performance of each rule
def evaluate_rule_performance(X, y, rules, le):
    per_rule_metrics = []
    total_instances = len(X)  # Total number of instances in the dataset

    for path, leaf_class in rules:
        # Create mask for rule
        mask = np.ones(len(X), dtype=bool)
        for condition in path:
            # Split the condition based on ' <=' or ' >'
            if ' <=' in condition:
                feature, value = condition.split(' <= ')
                operator = '<='
            elif ' >' in condition:
                feature, value = condition.split(' > ')
                operator = '>'
            else:
                feature, value = condition.split(' ', 1)
                mask &= (X[feature] == value)
                continue

            value = float(value)
            if operator == "<=":
                mask &= (X[feature] <= value)
            else:
                mask &= (X[feature] > value)

        # Subset data that falls under the rule
        X_subset = X[mask]
        y_subset = y[mask]

        if len(X_subset) == 0:
            continue  # Skip if no data points match this rule

        # Predicted labels (same size as y_subset)
        y_pred_subset = np.array([leaf_class] * len(y_subset))

        # Calculate metrics for this rule
        accuracy = accuracy_score(y_subset, y_pred_subset)
        precision = precision_score(y_subset, y_pred_subset, average='weighted', zero_division=1)
        recall = recall_score(y_subset, y_pred_subset, average='weighted', zero_division=1)
        f1 = f1_score(y_subset, y_pred_subset, average='weighted', zero_division=1)

        # Calculate rule simplicity (number of conditions in the rule)
        rule_simplicity = 1 / len(path)  # Inverse of number of conditions

        # Calculate coverage (percentage of data covered by this rule)
        coverage = len(X_subset) / total_instances

        per_rule_metrics.append({
            'rule': ' AND '.join(path),
            'class': leaf_class,  # Include the class for this rule
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'num_instances': len(X_subset),
            'simplicity': rule_simplicity,
            'coverage': coverage
        })

    return per_rule_metrics

# Evaluate per-rule performance
per_rule_metrics = evaluate_rule_performance(X, y, rules, le)

# Print per-rule performance
print("Per-Rule Performance:")
for rule_metrics in per_rule_metrics:
    print(f"Rule: {rule_metrics['rule']}")
    print(f"Class: {rule_metrics['class']}")  # Print the associated class
    print(f"Accuracy: {rule_metrics['accuracy']:.4f}")
    print(f"Precision: {rule_metrics['precision']:.4f}")
    print(f"Recall: {rule_metrics['recall']:.4f}")
    print(f"F1 Score: {rule_metrics['f1']:.4f}")
    print(f"Number of Instances: {rule_metrics['num_instances']}")
    print(f"Simplicity (1/Conditions): {rule_metrics['simplicity']:.4f}")
    print(f"Coverage: {rule_metrics['coverage']:.4f}")
    print("-" * 30)

# Compute the average performance across all rules
overall_accuracy = np.mean([m['accuracy'] for m in per_rule_metrics])
overall_precision = np.mean([m['precision'] for m in per_rule_metrics])
overall_recall = np.mean([m['recall'] for m in per_rule_metrics])
overall_f1 = np.mean([m['f1'] for m in per_rule_metrics])
overall_simplicity = np.mean([m['simplicity'] for m in per_rule_metrics])
overall_coverage = np.mean([m['coverage'] for m in per_rule_metrics])

print(f"Overall Accuracy: {overall_accuracy:.4f}")
print(f"Overall Precision: {overall_precision:.4f}")
print(f"Overall Recall: {overall_recall:.4f}")
print(f"Overall F1 Score: {overall_f1:.4f}")
print(f"Overall Simplicity: {overall_simplicity:.4f}")
print(f"Overall Coverage: {overall_coverage:.4f}")
```
